```python
#WAP to display all the lines of the file execpt thos that has # sign.
f1=open("task.txt","r")
f2=open("data.txt","w")
a=f1.readlines()
for i in a:
    if i[0]=="#":
        continue
    elif "#" in i:
        for j in range(len(i)):
            if i[j]=="#":
                f2.write(i[0:j-1:1] + "\n")
    else:
        f2.write(i)
f1.close()
f2.close()
```

```python
#WAP to compare two files and find the difference in line and character
f1=open("task.txt","r")
f2=open("data.txt","r")
s1=f1.readlines()
s2=f2.readlines()
for i in range(len(s1)):
    for j in range(len(s1[i])):
        if s1[i][j]==s1[i][j]:
            print("Line :",i+1)
            print("Index :",j+1)
            print(s1[i][j],"|",s2[i][j])
```
